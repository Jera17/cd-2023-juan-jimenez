```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import colors
from scipy.stats import pearsonr, zscore

import statsmodels.api as sm
import statsmodels.formula.api as smf

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import linear_model
```

```python
#Datos de propiedades
ruta = 'https://raw.githubusercontent.com/rasief/cartografia/master/melb_data.csv'
df = pd.read_csv(ruta)
```

```python
df = df.select_dtypes(exclude=['object'])
```

```python
df = df.dropna()
```

```python
#Remover valores más allá de 3 desviaciones entándar
df1 = df.copy()
for columna in df1:
  df1 = df1[(np.abs(zscore(df1[columna]))<=3)]
```

```python
df1.dtypes
```

#Cada X vs Precio

```python
# Sin incluir las columnas Postcode, Lattitude, Longtitude
max = 0
for i in df1.iloc[:,[0,2,3,4,5,6,7,8,9,12]]:
    print('/'*100)
    print('{} VS Price'.format(i))
    #Graficas
    sb.regplot(x=i, y="Price", data=df1, marker='+', x_jitter=0.2, y_jitter=0.2, line_kws={"color": "red"})
    plt.show()
    #OLS Regression
    x = df1[i]
    y = df1["Price"]
    x2 = sm.add_constant(x)
    est = sm.OLS(y, x2)
    est2 = est.fit()
    if max < est2.rsquared:
      max = est2.rsquared
      columnas = i
    print(est2.summary())
print('El mejor r^2 es {} vs Price: {}'.format(columnas,max))
```

#Pares de X vs Price

```python
# Sin incluir las columnas Postcode, Lattitude, Longtitude
aux = set()
max = 0
for i in df1.iloc[:,[0,2,3,4,5,6,7,8,9,12]]:
  aux.add(i)
  for j in df1.iloc[:,[0,2,3,4,5,6,7,8,9,12]]:
    if (i != j and not(j in aux)):
      #OLS Regression
      x = df1.loc[:,[i,j]]
      y = df1["Price"]
      x2 = sm.add_constant(x)
      est = sm.OLS(y, x2)
      est2 = est.fit()
      print(est2.summary())
      if max < est2.rsquared:
        max = est2.rsquared
        columnas = i + ' ' + j
print('El mejor r^2 es {} vs Price: {}'.format(columnas,max))
```

#Trios de X vs Price

```python
# Sin incluir las columnas Postcode, Lattitude, Longtitude
aux = set()
max = 0
for i in df1.iloc[:,[0,2,3,4,5,6,7,8,9,12]]:
  aux2 = set()
  aux.add(i)
  for j in df1.iloc[:,[0,2,3,4,5,6,7,8,9,12]]:
    aux2.add(j)
    for k in df1.iloc[:,[0,2,3,4,5,6,7,8,9,12]]:
      if (i != j and j != k and k != i and not(k in aux2) and not(j in aux)):
        #OLS Regression
        x = df1.loc[:,[i,j,k]]
        y = df1["Price"]
        x2 = sm.add_constant(x)
        est = sm.OLS(y, x2)
        est2 = est.fit()
        print(est2.summary())
        if max < est2.rsquared:
          max = est2.rsquared
          columnas = i + ' ' + j + ' ' + k
print('El mejor r^2 es {} vs Price: {}'.format(columnas,max))
```
